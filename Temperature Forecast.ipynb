```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

### I started by importing necessaries libraries including pandas for data manipulation and scikit-learn for modeling

```python
data= pd.read_csv(r"C:\Users\Dell\Downloads\temperature.csv")
data
```

```python
column_names = data.columns
for column in column_names:
        print(column + ' - ' + str(data[column].isnull().sum()))
```

```python
data = data.drop(columns=['Date'], axis=1)
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

### checked missing data in the dataset by iterating through the column names and printing the count of null values for each column

```python
X = data.drop(['Next_Tmax', 'Next_Tmin'], axis=1)
y_max = data['Next_Tmax']
y_min = data['Next_Tmin']
```

```python
X_train_max, X_test_max, y_train_max, y_test_max = train_test_split(X, y_max, test_size=0.2, random_state=42)
X_train_min, X_test_min, y_train_min, y_test_min = train_test_split(X, y_min, test_size=0.2, random_state=42)
```

```python
model_max = LinearRegression()
model_max.fit(X_train_max, y_train_max)
```

```python
model_max.score(X_train_max, y_train_max)
```

```python
y_pred_max = model_max.predict(X_test_max)
mae_max = mean_absolute_error(y_test_max, y_pred_max)
```

```python
y_pred_min = model_min.predict(X_test_min)
mae_min = mean_absolute_error(y_test_min, y_pred_min)
```

```python
print(f'MAE for Next_Tmax: {mae_max}')
print(f'MAE for Next_Tmin: {mae_min}')
```

## The MAE for Next_Tmax is approximately 1.1177, which represents the average absolute difference between the predicted and actual values for Next_Tmax
